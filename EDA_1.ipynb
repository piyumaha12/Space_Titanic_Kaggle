```python
def Label_encoding(data_original, columns_name = None, starting_col= None, end_col = None):
    # Extracting obj types
    
    encoding_dict_total = {}
    data = data_original.copy()
    if columns_name:
        data_1 = data[columns_name]
        list_of_col = columns_name
    else:
        data_1 = data.iloc[:,starting_col:end_col+1]
        list_of_col = data_1.dtypes[data.dtypes == object].index
    for col in list_of_col:
        labels = list(data_1[col].unique())
        if type(labels[-1])== float:
            # print('in if')
            labels= labels[:-1]
        range_of_encoding = range(len(labels))
        encoding_dict = dict(zip(labels, range_of_encoding))
        encoding_dict_total[col] = encoding_dict
        print(encoding_dict)
        data[col] = data[col].map(encoding_dict)
    return encoding_dict_total, data
    
```

```python
file_dict, data_space = Label_encoding(space_df, columns_name= ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP'])
```
